```
from torchviz import make_dot
import torch

x=torch.ones(2, requires_grad=True)
y=2*x
z=3+x
r=(y+z).sum()    
make_dot(r)
```

```
from torchviz import make_dot
import torch

x=torch.ones(2, requires_grad=True)
y=2*x
z=3+x.detach()
r=(y+z).sum()    
make_dot(r)
```

```
from torchviz import make_dot
import torch

x=torch.ones(2, requires_grad=True)
y=2*x
z=3+x.data # .data is old method, detach is new
r=(y+z).sum()    
make_dot(r)
```
